```python
#!/usr/bin/env python

import urllib2
import re
import collections
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import synapseclient

def loadftp(link):
    req = urllib2.Request(link)
    resp = urllib2.urlopen(req)
    dat = resp.readlines()
    return(dat)

def getSub(dat, a, b):
    return(dat[a:b])

def getItem(dat, item):
    out = [s.split()[1] for s in dat if s.startswith(item)]
    return("".join(out))
    
def getNormalChr(dat):
    out = {s.split()[1]:s.split()[3] for s in dat if s.startswith("NormalChromosome")}
    for k in out.keys():
        if "X" in k:
            newk = 23
            out[newk] = out.pop(k)
        if "Y" in k:
            newk = 24
            out[newk] = out.pop(k)
    out={int(k):int(v) for k, v in out.iteritems()}
    return(out)

def getFrags(dat):
    out = {}
    for s in dat:
        if s.startswith("SkyFrag"):
            frags = s.split()[2:]
            if frags[0] in out.keys():
                out[frags[0]].append(frags[1:])
            else:
                out[frags[0]]=[frags[1:]]
    for k in out.keys():
        if "X" in k:
            newk = 23
            out[newk] = out.pop(k)
        if "Y" in k:
            newk = 24
            out[newk] = out.pop(k)
        if "U" in k:
            newk = 0
            out[newk] = out.pop(k)
    out={int(k):v for k, v in out.iteritems()}
    return(out)

def cumSum(x):
    cum = [0]
    for i in range(len(x)):
        cum.extend([cum[i]+x[i]])
    return(cum)
   
def getChrLen(synId="syn2141399"):
    '''Dowload hg19 Chr length from synapse'''
    ent = syn.get(synId)
    hg = [l.split() for l in open(ent.path, 'rU')][1:]
    hg = {int(l[0]):[float(l[i]) for i in range(1,4)] for l in hg}
    return(hg)

def getCytoBand(synId="syn2319170"):
    '''Dowload hg19 cytoband locations from synapse'''
    ent = syn.get(synId)
    hg = [l.split() for l in open(ent.path, 'rU')][1:]
    hg = {(re.sub("chr", "",h[0]),h[3]):(h[1],h[2]) for h in hg}
    for k in hg.keys():
        if "X" in k:
            newk = (23, k[1])
            hg[newk] = hg.pop(k)
        if "Y" in k:
            newk = (24, k[1])
            hg[newk] = hg.pop(k)
    hg = {(int(k[0]), k[1]):v for k, v in hg.iteritems()}
    return(hg)

def mapCytoband(CytoBandDic, Chr, band):
    start=band[0]
    end=band[1]
    if start=="q10": start="cen"
    if end=="q10": end="cen"
    
    tmp={k[1]:CytoBand[k] for k in CytoBand if Chr in k}
    startLoc, endLoc = [], []

    if start=="?":
        startLoc.append(None)
    elif start=="pter":
        startLoc.append(0)
    elif start=="qter":
        startLoc.append(float(hg[Chr][0]))
    elif start=="cen":
        startLoc.append(sum(hg[Chr][1:])/2)
    else:
        for k in tmp:
            if re.search(start, str(k)):
                startLoc.append(float(tmp[k][0]))
    
    if end=="?":
        endLoc.append(None)
    elif end=="pter":
        endLoc.append(0)
    elif end=="qter":
        endLoc.append(float(hg[Chr][0]))
    elif end=="cen":
        endLoc.append(sum(hg[Chr][1:])/2)
    else:
        for k in tmp:
            if re.search(end, str(k)):
                endLoc.append(float(tmp[k][1]))
    Loc = pd.Series(startLoc+endLoc)
    return([Loc.min(skipna=True), Loc.max(skipna=True)])

def mapFrags(normal, frags, hg, CytoBandDic):
    out={Chr:[[0, hg[Chr][0]] for i in range(v)] for Chr, v in normal.iteritems()}
#    for Chr, V in out.iteritems():
#        l=[]
#        for v in V:
#            l.extend(v)
#        out[Chr]=l
    for Chr in frags:
        for band in frags[Chr]:
            band=band[:2]
            if Chr in out:
                out[Chr].append(mapCytoband(CytoBandDic, Chr, band))
#                out[Chr]=out[Chr]+mapCytoband(CytoBandDic, Chr, band)
            else:
                out[Chr]=mapCytoband(CytoBandDic, Chr, band)
    out={Chr:out[Chr] for Chr in out}
    return(out)

def reverse(x, y):
        tmp=x
        x=y
        y=tmp
        return x, y
    
def included(seg1, seg2):
    d1 = abs(seg1[0]-seg1[1])
    d2 = abs(seg2[0]-seg2[1])
    if d2>d1:
        seg1, seg2 = reverse(seg1, seg2)
    test=seg2[0]>=seg1[0] and seg2[1]<=seg1[1]
    return(test)

def allBreaks(Dict):
    frags=[k2 for k1 in Dict for k2 in k1[1:]]
    frags=sorted(set(frags))
    frags=[(frags[i], frags[i+1]) for i in range(len(frags)-1)]
    return(frags)
    
def getSegValues(breaks, segList):
    out={}
    for b in breaks:
        value=0
        for seg in segList:
            Chr=seg[0]
            if included(b, seg[1:]):
                value+=segList[seg]
        out[tuple([Chr])+b]=value
    return(out)

def getCNlike(frags):
    CNlike = {}
    for Chr in frags:
        tmp={(Chr,a,b):v for (a,b),v in collections.Counter(map(tuple, frags[Chr])).iteritems()}
        breaks=allBreaks(tmp)
        CNlike.update(getSegValues(breaks, tmp))
        print
    return(CNlike)

def convertGenomic(frags, genomicLoc):
    CNlike = getCNlike(frags)
    out={}
    for loc in CNlike:
        Chr, start, end = (loc[i] for i in range(len(loc)))
        start, end = start+genomicLoc[Chr][0], end+genomicLoc[Chr][0]
        out[(Chr, start, end)]=CNlike[loc]
    return(out)

def exportToCSV(CN, name, path="/Users/fredcommo/Documents/MyProjects/CellLines/NCI60/Karyotypes/Converted_Karyotypes/", extend=".tsv"):
    name=name.replace("/", "_")
    Chr=[k[0] for k in CN]
    x1=[k[1] for k in CN]
    x2 =[k[2] for k in CN]
    y=[v for v in CN.values()]
    X = np.vstack((Chr, x1, x2, y)).T
    X = pd.DataFrame(X, columns = ["Chr", "start", "end", "value"])
    X = X.sort(["Chr", "start", "end"])
    X.to_csv(path+name+extend, sep = "\t", index = False)

```

```python
hg = getChrLen()
cum = cumSum([l[0] for l in hg.values()])
genomicLoc = {i:[cum[i-1]+1, cum[i]] for i in range(1, 25)}
```

```python
# Get skyCases (cell lines) indices.
skyCases = {i:re.sub("\\((.*)\\)", " ", s).split()[1:] for i, s in enumerate(dat) if s.startswith('SkyCase')}
idx = sorted(skyCases.keys())
idx.extend([len(dat)])
for v in skyCases.values():
	while '"' in v:
		v.remove('"')
skyCases = {v[0].replace('"', ""):v[1:] for v in skyCases.values()}
#print skyCases
print idx
print len(idx)
```

```python
for i in range(len(idx)-1):
    if i != 51:
        cell = getSub(dat, idx[i], idx[i+1])
        normal=getNormalChr(cell)
        frags = getFrags(cell)
        frags=mapFrags(normal, frags, hg, CytoBand)
        CN=convertGenomic(frags, genomicLoc)
        exportToCSV(CN, name=getItem(cell, "casename"))
        print i, "Cell line", getItem(cell, "casename")
```
